Major differences between Supervised learning, Unsupervised learning and Reinforcement Learning is that the Supervised learning is based on prefeeding the machine with several data sets and helping it to distinguish between those under supervision of the trainer while in unsupervised learning the machiine differentiates between the datasets on its own but reinforcement learning has no pre data set and works on the principle of rewarding for correct answer and punishing for wrong answer.
Among these 3 learning tactics supervised learning is better than unsupervised because it is more accurate, labelled and easier to apply. It is more applicable in places like speech recognition, spam detection for mails.
Also Reinforcement learning is better than Supervised learning because it can also correct the mistakes done during the training process. It is beneficial in applications like chess engines where each output depends on the state of the current input.
Unsupervised learning is also useful when the datasets given are easier to manipulate and understand automatically like in E-commerce websites which detect the customer purchases and then gives suggestions based on it.

Difference between Machine learning and Deep learning is that the data used is Machine learning is quite structured while that in Deep learing is instead like a neural network.
Machine learning consists of thousands of data points while deep learning consists of millions of datapoints. Machine learning is widely used in areas like Spotify, Youtube, etc.

**Write a function to check whether the number is Palindrome or not.**  
- Hopefully who have did this in c++ in yor CS101

```python
n=int(input("enter a number: "))
word=str(n)
letters=list(word)
rletters=letters[::-1]
if letters == rletters:
    print(n,"is a palindrome")
else:
    print(n,"is not a palindrome")
```


Create a NumPy array of 10 random integers between 1 and 100 and calculate the mean, minimum, and maximum values


```python
import numpy as np
import random
A = []
for i in range(10):
    A.append(random.randint(1,100))
arr=np.array(A)
print(arr)
print(arr.mean())
print(arr.min())
print(arr.max())
```
